```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
                                    #Modules Imported
data=yf.Ticker('BTC-USD').history(start='2021-12-01',end='2025-01-01',interval='1d')                                   #extra 30 days for ROC to get into action
data['ROC'] = data['Close'].pct_change(periods=14) * 100
data = data.dropna()                                                                                                                 #ROC
data['MFM']=((data['Close'] - data['Low'])-(data['High'] - data['Close'])) / (data['High'] - data['Low']) 
data['MFM']=data['MFM'].fillna(0) 
data['MFV']=data['MFM']*data['Volume']
data['AD']=data['MFV'].cumsum()                                                                                                     #A/D LINE
vix=yf.download("^VIX", start='2021-01-01', end='2025-01-01')                                                                        # VIX
data.index = data.index.tz_localize(None)                                                                              #To localise timezone for VIX and BTC-USD
vix.index = vix.index.tz_localize(None)
data['VIX'] = vix['Close'].reindex(data.index,method='ffill')  
data['SMA200']=data['Close'].rolling(window=200).mean()
data['SMA100']=data['Close'].rolling(window=100).mean()
                                                       #Matching Dates And Filling Values For NaN
                                 #Indicator Coded
data['Signal'] = 0
data['AD2']=data['AD'].diff()                                                                                              #To Detremine UP and DOWN of A/D LiNE
data.loc[(data['ROC'] > 0) &(data['AD2'] > 0 )&(data['VIX']>15)&(data['VIX']<22)&(data['SMA100']<data['SMA200']) ,'Signal'] = 1 
data.loc[(data['ROC'] < -0.5) &(data['AD2'] < 0)&(data['VIX']>15)&(data['VIX']<22) &(data['SMA100']>data['SMA200']),'Signal'] = -1
                                    #Signal Generated
data['Marketreturn']=data['Close'].pct_change()
data['VOL'] = data['Marketreturn'].rolling(14).std() #used position sizing 
data['VOL_target'] = data['VOL'].rolling(30).mean()
data['PositionSize'] = data['VOL_target'] / data['VOL']
data['PositionSize'] = data['PositionSize'].clip(lower=0, upper=1)
data['PositionSize'] = data['PositionSize'].fillna(0)                                                                                       #CHANGE   
data['Strategyreturn']=data['Marketreturn']*data['Signal'].shift(1)* data['PositionSize'].shift(1)                                                 #CHANGE ACCORDING TO MY STRATEGY
data['CUMmarket']=(1+data['Marketreturn']).cumprod()                                                                         #PORTFOLIO CHANGE ON B&H
data['CUMstrategy']=(1+data['Strategyreturn']).cumprod()                                                                  #PORTFOLIO CHANGE ON STRATEGY
data[['CUMmarket','CUMstrategy']] = data[['CUMmarket','CUMstrategy']].fillna(1)                                                     # NaN SORTED
                                      #Backtesting Done
mp.figure(figsize=(12,6))
mp.plot(data.index,data['CUMmarket'],label='BUY&HOLD')
mp.plot(data.index,data['CUMstrategy'],label='STRATEGY')
mp.legend()
mp.title('BTC-USD')
mp.grid(True) 
mp.show()
                                         #Graph Plotted
print('TOTAL PROFIT :',(data['CUMstrategy'].values[-1]-1)*100)
print('NET PROFIT :',(data['CUMstrategy'].values[-1]-1)*100)
data['Change'] = data['Signal'].diff()
entryp = data['Close'].where(data['Change'] != 0)
exitp = entryp.shift(-1)
direction = data['Signal'].where(data['Change'] != 0)
returns= direction * (exitp - entryp) / entryp
returns = returns.dropna()
winningtrades = (returns > 0).sum()
losingtrades = (returns < 0).sum()
totaltrades = len(returns)
print("TOTAL TRADES :", totaltrades)
print("WINNING TRADES :", winningtrades)
print("LOSING TRADES :", losingtrades)
data['PEAK']=data['CUMstrategy'].cummax()
data['Drawdown']=(data['CUMstrategy']-data['PEAK'])/data['PEAK']
print('MAX DRAWDOWN :',-data['Drawdown'].min()*100)
avg=data['Strategyreturn'].mean()
risk=data['Strategyreturn'].std()
print('SHARPE RATIO :',(avg/risk)* np.sqrt(252))
data.loc[data['Strategyreturn']<0,'Sort']=data['Strategyreturn']
sortrisk=data['Sort'].std()
print("SORTINO RATIO :",(avg/sortrisk)* np.sqrt(252))

```

```python
#MOMENTUM INDICATOR : RATE OF CHANGE (ROC)
import yfinance as yf
import pandas as pd
from datetime import datetime,timedelta
import matplotlib.pyplot as mp
ticker="AAPL"
n=30
enddate=datetime.today()    #for current date
startdate=enddate-timedelta(days=3*n)    #for avoiding NIN values in dataframes
df=pd.DataFrame()
data=yf.download(ticker,start=startdate,end=enddate) # usinf y finance to download 
df['Close']=data['Close']
df['ROC']=((df['Close']-df['Close'].shift(n))/df['Close'].shift(n))*100   #main formula
df=df.dropna() #to avoid initial ROC NIN value for clear plot
mp.figure(figsize=(12,6)) #canvas
mp.plot(df.index,df['ROC'],linewidth=2) #ploting axis 
mp.axhline(0,color='black',linestyle='--') # defing zero
maxpnt=df['ROC'].abs().max() #considering negative values also
mp.ylim(-maxpnt,maxpnt) #as ROC may be negative
mp.title(f"ROC INDICATOR FOR MOMENTUM FOR {ticker}") # titling
mp.xlabel("DATE")
mp.ylabel('ROC')
mp.grid(True) # to show better graph
mp.show()
```

```python
import yfinance as yf
import pandas as pd 
import matplotlib.pyplot as mp
from datetime import datetime,timedelta 
ticker="AAPL"
n=5
enddate=datetime.today() #for current date
startdate=enddate-timedelta(days=3*n) #for avoiding non trading days 
data=yf.Ticker(ticker).history(start=startdate,end=enddate) #to avoid mltilevel 
data['MFM']=((data['Close'] - data['Low'])-(data['High'] - data['Close'])) / (data['High'] - data['Low']) #formula
data['MFM']=data['MFM'].fillna(0) #to compensate where high = low
data['MFV']=data['MFM']*data['Volume']#formula
data['AD']=data['MFV'].cumsum()#formula
mp.figure(figsize=(12,6))
mp.plot(data.index,data['AD'],linewidth=2)
mp.title(f'A/D LINE for {ticker}')
mp.xlabel('DATE')
mp.ylabel("A/D (Cumulative Volume)")
mp.axhline(0,color='black', linestyle='--')
maxpnt=data['AD'].abs().max()
mp.ylim(-maxpnt,maxpnt)#to have clear graph
mp.grid(True)
mp.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as mp
import pandas as pd
startdate=input('STARTDATE (yyyy-mm-dd) : ')
enddate=input('ENDDATE (yyyy-mm-dd) : ')
data=yf.Ticker('AAPL').history(start=startdate,end=enddate) #To fix multi column issue in newer yfinance version
body=(data['Close']-data['Open']).abs()
upshadbody=data['High']-data[['Close','Open']].max(axis=1) #Defined Bodies
lowshadbody=data[['Close','Open']].min(axis=1)-data['Low']
data['hammer']=0
data.loc[(upshadbody<body*0.3)&(lowshadbody>=2*body)&(body>0),'hammer']=1 #defined cinditionn for CANDLE
print("No. of Hammers = ",data['hammer'].sum())

hammmer=data[data['hammer']==1]
mp.figure(figsize=(12,6))
mp.plot(data.index,data['Close'],linewidth=2,)
mp.scatter(hammmer.index,hammmer['Close'],color='red',marker='^',s=100,label='Hammer Candles') # used to define conditional points of the CANDLE
mp.xlabel('Date')
mp.ylabel('Price')
mp.legend()
mp.grid(True)
mp.show()
```
